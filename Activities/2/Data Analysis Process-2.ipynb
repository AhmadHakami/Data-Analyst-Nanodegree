# Wrangle the dataset (clean it)

```python
data['Salary'] = data['Salary'].fillna(data['Salary'].mean())
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())
```

```python
data['Age'] = data['Age'].astype(int)
```

```python
data.sort_values(by = ['Salary'], ascending = False, inplace = True)
```

```python
data
```

```python
salaries = data.groupby('Country')['Salary'].mean().sort_values(ascending = False)
```

```python
salaries.plot(kind = 'bar'); 
```

```python
educations = data.groupby('Education')['Salary'].mean().sort_values(ascending = False)
```

```python
educations.plot(kind = 'bar');
```
